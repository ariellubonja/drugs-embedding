```python
from langchain.embeddings import LlamaCppEmbeddings
import numpy as np
import pandas as pd
```

```python
llama2 = LlamaCppEmbeddings(model_path="./models/llama-2-7b.gguf.q8_0.bin")
llama = LlamaCppEmbeddings(model_path="./models/llama-7b.gguf.q8_0.bin")
```

```python
original_df = pd.read_csv('data/BioClinicalBERT_metadata.tsv', index_col=None, sep='\t')
original_df.head()
```

```python
%%capture

# Llama2 embeddings
# doc_result = llama2.embed_documents(list(original_df['Molecule']))
# Llama1
llama1_embeddings = llama.embed_documents(list(original_df['Molecule']))
```

```python
np.array(llama1_embeddings).shape
```

```python
# llama2_embeddings = pd.DataFrame(doc_result, index=original_df['Molecule'])
llama1_embeddings = pd.DataFrame(llama1_embeddings, index=original_df['Molecule'])
llama1_embeddings
```

```python
# llama2_embeddings.to_csv('data/Llama2_embeddings.tsv', sep='\t', index=False, header=False)
llama1_embeddings.to_csv('data/Llama1_embeddings.tsv', sep='\t', index=False, header=False)
```
