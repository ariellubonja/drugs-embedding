## So far we fetched all drugs starting with Ab

Now get all that start with A

```python
import requests
from bs4 import BeautifulSoup

# Note this page is different from the one above. This is A. Above is Ab
url_a = "https://www.drugs.com/alpha/a.html"

parsed_page = BeautifulSoup(requests.get(url_a).text)
```

```python
parsed_page
```

```python
# Find this to get the possible links for the letter A, e.g. Ab, Ac, Ad...

# h1>Drugs: A</h1><p>List of drugs and medications that start with the letter '<b>A</b>'.</p><nav aria-label="Drug list navigation by first letter" class="ddc-paging paging-list-wrap ddc-mgb-2">
# <ul>
# <li class=""><span>Aa</span></li>
# <li><a aria-label="Browse drugs starting with: Ab" href="/alpha/ab.html">Ab</a></li>
# <li><a aria-label="Browse drugs starting with: Ac" href="/alpha/ac.html">Ac</a></li>
# <li><a aria-label="Browse drugs starting with: Ad" href="/alpha/ad.html">Ad</a></li>
# <li><a aria-label="Browse drugs starting with: Ae" href="/alpha/ae.html">Ae</a></li>

# Find first <ul> with no class name
# https://www.scrapingbee.com/webscraping-questions/beautifulsoup/how-to-find-elements-without-specific-attributes-in-beautifulsoup/
ul_all_no_class = parsed_page.find("body").findAll("ul",attrs={"class": None})
ul_all_no_class
```

```python
len(ul_all_no_class)
```

```python
# TODO this might not be general, may have to fix if pages aren't consistent

# Selects the grid of links that specifies the first two letters of the drug name
#   e.g. Aa, Ab, Ac. Not all of them are valid e.g. Aa doesn't exist in this case
ul_first_two_letters = ul_all_no_class[2]
ul_first_two_letters
```

```python
two_letter_drug_links = [a['href'] for a in ul_first_two_letters.findAll("a")]
two_letter_drug_links
```

```python
base_url = "https://www.drugs.com"
base_url + two_letter_drug_links[0]
```

```python
import urllib, time, random

# Add a random wait time of .5-1s for each request, to avoid getting banned
def crawl_and_wait(list_of_urls):
    i = 1
    for url in list_of_urls:
        # Cannot create file names with same name as href, since it contains slashes
        urllib.request.urlretrieve(url, url.replace("/","$"))
        time.sleep(random.uniform(0.5,1))
        i += 1

        if i % 20 == 0:
            print("Progress: ", i, " / ", len(list_of_urls))
```

```python
full_two_letter_urls = [base_url + i for i in two_letter_drug_links]
full_two_letter_urls
```

```python
all_a_drug_urls = [html_tools.get_drug_urls_in_list_page(url) for url in full_two_letter_urls]
all_a_drug_urls
```

```python
# Flatten list of lists
# https://stackoverflow.com/questions/952914/how-do-i-make-a-flat-list-out-of-a-list-of-lists
flat_a_drugs_list = [item for sublist in all_a_drug_urls for item in sublist]
print(flat_a_drugs_list)
print(len(flat_a_drugs_list))
```

### Let's do the above for the other drugs B-Z

Create function for getting all drug links under "letter"

```python
import requests
from bs4 import BeautifulSoup


# Gets the links to all specific drugs from a certain letter e.g. gets URLs of all A-drugs
def fetch_all_drug_links_starting_with(letter):
    time.sleep(random.uniform(0.5,1))

    url_a = "https://www.drugs.com/alpha/"+ letter + ".html"
    parsed_page = BeautifulSoup(requests.get(url_a).text)
    ul_all_no_class = parsed_page.find("body").findAll("ul",attrs={"class": None})
    ul_first_two_letters = ul_all_no_class[2]
    two_letter_drug_links = [a['href'] for a in ul_first_two_letters.findAll("a")]

    base_url = "https://www.drugs.com"
    full_two_letter_urls = [base_url + i for i in two_letter_drug_links]
    all_a_drug_urls = [get_drug_urls_in_list_page(url) for url in full_two_letter_urls]
    flat_a_drugs_list = [item for sublist in all_a_drug_urls for item in sublist]

    return flat_a_drugs_list
```

```python
alphabet = ["a", "b", "c", 'd','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

```python
# Except ones skipped due to different page layout - about 15%
most_drug_links = [fetch_all_drug_links_starting_with(l) for l in alphabet]
most_drug_links
```
